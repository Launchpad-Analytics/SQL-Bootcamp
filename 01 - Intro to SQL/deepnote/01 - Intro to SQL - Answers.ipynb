### Exercise 7

### Exercise 8

### Exercise 9

Calculate the total populations for each county in New York

```python


df_10 = _deepnote_execute_sql("""SELECT county_name, SUM(population) as total_population
FROM Cities
WHERE state_id = 'NY'
GROUP BY county_name
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_10
```

### Exercise 10

Find the `county_name`, `state_name` and total population for all counties in the states of New Jersey, Pennsylvania, New York, and Delaware

```python


df_11 = _deepnote_execute_sql("""SELECT county_name, state_name, SUM(population)
FROM Cities
WHERE state_id IN ('NJ', 'PA', 'NY', 'DE')
GROUP BY county_name, state_name
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_11
```

### Exercise 11

Find the top 5 largest countied in New Jersey based on total population.

```python


df_12 = _deepnote_execute_sql("""SELECT county_name, SUM(population) as total_population
FROM Cities
WHERE state_name = 'New Jersey'
GROUP BY county_name
ORDER BY total_population DESC
LIMIT 5
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_12
```

## Exercise 12

Generate a table of all counties and their respective states that contain more than 50 cities.

```python


df_13 = _deepnote_execute_sql("""SELECT county_name, state_name, count(*) as num_cities
FROM Cities
GROUP BY county_name, state_name
HAVING num_cities > 50
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_13
```
